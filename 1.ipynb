```python
# Import the necessary libraries
import pandas as pd

```

```python
# Import the dataset from the given address
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
```

```python
# Discover what is the mean age per occupation
occupation_by_age = users.groupby("occupation")["age"].mean()
occupation_by_age
```

```python
# Discover the Male ratio per occupation and sort it from the most to the least
male_ratio = users.groupby("occupation")["gender"].apply(lambda x: (x == "M").mean()).sort_values(ascending=False)
male_ratio
```

```python
# For each occupation, calculate the minimum and maximum ages
ages_per_occupation = users.groupby("occupation")["age"].agg(['min', 'max'])
ages_per_occupation['male_ratio'] = male_ratio
ages_per_occupation
```

```python
# For each combination of occupation and sex, calculate the mean age
mean_age = users.groupby(["occupation", "gender"])["age"].mean()
mean_age
```

```python
# For each occupation present the percentage of women and men
mean_age_df = mean_age.reset_index()
gender_percent_df['mean_age'] = mean_age_df['age']
gender_percent_df
```
