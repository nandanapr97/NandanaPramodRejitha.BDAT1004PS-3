```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
url = 'https://raw.githubusercontent.com/nandanapr97/NandanaPramodRejitha.BDAT1004PS-3/main/us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(url)
data
```

```python
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per Capita', marker='o')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per Capita', marker='o')

#Label the plot
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Number per 1000 People')
plt.legend()

# Show the plot
plt.show()

```
