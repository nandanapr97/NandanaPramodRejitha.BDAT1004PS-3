```python
#Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Import the dataset from the attached file wind.txt
# Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

# URL for my data
url = 'https://raw.githubusercontent.com/nandanapr97/NandanaPramodRejitha.BDAT1004PS-3/main/wind.txt'

# Read the data using parse_dates to combine the first three columns into a datetime index
data = pd.read_csv(url, delim_whitespace=True, header=None, na_values='NaN', parse_dates=[[0, 1, 2]])

# Rename columns
data.columns = ['Date', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']

# Set the 'Date' column as the index
data = data.set_index('Date')

data.head()
```

```python
#Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

# Function to fix the year
def fix_year(x):
    if x.year > 1989:
        year = x.year - 100
    else:
        year = x.year
    return pd.Timestamp(year=x.year, month=x.month, day=x.day)

# Apply the function to the 'Date' column
data.index = data.index.map(fix_year)

# Display the updated DataFrame
data.head()
```

```python
#Set the right dates as the index.

# Convert 'Date' to datetime
data.index = pd.to_datetime(data.index)
data.head()

```

```python
#Compute how many values are missing for each location over the entire record. They should be ignored in all calculations below.

# Count missing values for each location
missing_values = data.isnull().sum()
missing_values
```

```python
#Compute how many non-missing values there are in total.


non_missing_values = data.notnull().sum().sum()
non_missing_values
```

```python
#Calculate the mean windspeeds of the windspeeds over all the locations and all the times. A single number for the entire dataset.

mean_windspeed = data.mean().mean()
mean_windspeed
```

```python
# Create a DataFrame called loc_stats and calculate the min, max, and mean windspeeds and standard deviations of the windspeeds at each location over all the days. A different set of numbers for each location.

# Calculate location statistics
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

loc_stats
```

```python
#. Create a DataFrame called day_stats and calculate the min, max, mean, and standard deviations of the windspeeds at each day. A different set of numbers for each day.

# Calculate daily statistics
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})
day_stats
```

```python
#Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.


avg__Jan_windspeed = data[data.index.month == 1].mean()
avg__Jan_windspeed
```
