```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://raw.githubusercontent.com/nandanapr97/NandanaPramodRejitha.BDAT1004PS-3/main/actor_kill_counts.csv'
actors_data = pd.read_csv(url)

print(actors_data.head())


```

```python
# Sort the dataset by kill count
actors_by_kill_count = actors_data.sort_values(by='Count', ascending=True)

# Create a bar chart
plt.figure(figsize=(10, 8))
plt.barh(actors_by_kill_count['Actor'], actors_by_kill_count['Count'], color='purple')
plt.xlabel('Kill Count')
plt.ylabel('Actors')
plt.title('Comparison of deadliest Actors in Hollywood')
plt.show()
```
