```python
#  Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Create 3 different Series
np.random.seed(42)  
series1 = pd.Series(np.random.randint(1, 5, size=100))
series2 = pd.Series(np.random.randint(1, 4, size=100))
series3 = pd.Series(np.random.randint(10000, 30001, size=100))

print(series1, series2, series3)
```

```python
# Create a DataFrame by joining the Series by column
df = pd.concat([series1, series2, series3], axis=1)
df
```

```python
# Change the name of the columns
df.columns = ['bedrs', 'bathrs', 'price_sqr_meter']
df
```

```python
# Create a one-column DataFrame with the values of the 3 Series
one_column_df = pd.DataFrame(pd.concat([series1, series2, series3], axis=0))
one_column_df
```

```python
#Ops it seems it is going only until index 99. Is it true?
print(one_column_df.index.max() == 99)
```

```python
#Reindex the DataFrame so it goes from 0 to 299
one_column_df.reset_index(drop=True, inplace=True)
one_column_df
```
