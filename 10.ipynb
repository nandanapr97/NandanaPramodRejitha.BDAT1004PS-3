```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the data
url = 'https://raw.githubusercontent.com/nandanapr97/NandanaPramodRejitha.BDAT1004PS-3/main/arcade-revenue-vs-cs-doctorates.csv'
data = pd.read_csv(url)
data


```

```python
# Scatter plot  based on the 'Year' column
plt.figure(figsize=(10, 8))
scatter = plt.scatter(
    data['Total Arcade Revenue (billions)'],
    data['Computer Science Doctorates Awarded (US)'],
    c=data['Year'],
    cmap='plasma',  
    alpha=0.9, 
    edgecolors='m', 
    linewidths=0.9  
)

# Labels and title
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Number of Computer Science PhDs Awarded (US)')
plt.title('Relationship between Arcade Revenue and Computer Science PhDs (2000-2009)')

# Legend for the color map
legend = plt.colorbar(scatter, label='Year')

# Show the plot
plt.show()
```
